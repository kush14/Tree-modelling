From the above correlation table,we observe that 'Fhi' and 'NHR' are having less correlation(-0.166136 and 0.189429 respectively) with respect to status . So, we can drop these two features based on their correlations.

```python
x=disease_df.drop(['MDVP:Fhi(Hz)','NHR','status','name'],axis=1)
```

```python
y=disease_df['status']
```

```python
#Checking Balance of 'status' column and we can see that it's highly imbalanced 
#almost 1:3 in favor of status 1. So, the model's ability to predict status 1 will be better than predicting status 0
disease_df['status'].value_counts().sort_index()
```

```python
#Splitting the data into train and test in 70/30 ratio with random state as 1
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import preprocessing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
#Decision tree modelling
dtree_model = DecisionTreeClassifier(criterion = 'entropy' )
dtree_model.fit(x_train, y_train )
```

Decision tree Model Accuracy

```python
#Decision tree model accuracy on test data as 86.44% which is without regularization
dtree_model.score(x_test , y_test)
```

```python
dtree_model.score(x_train , y_train)
```

```python
y_pred = dtree_model.predict(x_test)
```

```python
y_cm=metrics.confusion_matrix(y_test,y_pred)
```

Confusion matrix for Decision tree

```python
print(y_cm)
```

Decision tree with regularization

```python
dtree_model = DecisionTreeClassifier(criterion = 'entropy', class_weight={0:.5,1:.5}, max_depth = 5, min_samples_leaf=5 )
dtree_model.fit(x_train, y_train)
```

```python
print(dtree_model.score(x_train , y_train))

print(dtree_model.score(x_test , y_test))
```

```python
#Accuracy as 89.8% when DT is regularized with max depth and min sample leaf which is better than normal decision tree model
```

Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier(n_estimators = 50)
rfcl = rfcl.fit(x_train, y_train)
```

```python
y_predict = rfcl.predict(x_test)
print(rfcl.score(x_test , y_test))
print(metrics.confusion_matrix(y_test, y_predict))
```

```python
#Accuracy as 81.3% which is less than Decision tree as well as regularized DT
#so it seems that Regularized DT is the best model here
```
